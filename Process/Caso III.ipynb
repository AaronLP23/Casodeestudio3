```python
# Se importa el archivo y de determina el delimitador (;)
```

```python
df=pd.read_csv(inputpath,delimiter=";")
```

```python
# Se eliminan las columnas que tienen más de un 70% de valores nulos.
```

```python
for x in df:
    y= df[x].isna().mean()
    if y>0.7:
        df.drop(x, inplace=True,axis=1)
   
```

```python
# Se eliminan las filas que tienen menos de 2 elementos.
```

```python
cols=df.columns

for i in range(0, len(df)):
    y=df.loc[i,cols].isna().sum()
    if y>2:
        df=df.drop([i])
```

```python
df
```

```python
# Se limpia/procesa el dataset resultante.
```

```python
df.replace(to_replace="--", value=np.nan, inplace=True)
```

```python
df["Gender"].replace(to_replace=np.nan,value="Unknown",inplace=True)
```

```python
df["Department"].replace(to_replace=np.nan,value="Bench",inplace=True)
```

```python
df["HourlyRate"].replace(to_replace=np.nan,value=df["HourlyRate"].median(),inplace=True)
```

```python
df["OverTime"].replace(to_replace=[np.nan,"0"],value="No",inplace=True)
```

```python
df["OverTime"].replace(to_replace=["1"],value="Yes",inplace=True)
```

```python
# Se guarda el df al final del procesamiento indicado.
```

```python
df.to_csv(outputpath)
```
